```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
dir(keras)
```

```python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.mnist.load_data()
```

```python
x_train[0]
```

```python
len(x_train)
```

```python
y_train[0]
len(x_train)
len(x_train[0])
```

```python
plt.matshow(x_train[0])
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[1:3]
```

```python
len(x_train[0].flatten())
x_train[0].flatten()
```

```python
x_train_falttened=x_train.reshape(len(x_train),784)
x_test_falttened=x_test.reshape(len(x_test),784)
x_train_falttened
```

```python
len(x_train_falttened)
```

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation="sigmoid")
])
model.compile(
optimizer='adam',
    loss="sparse_categorical_crossentropy",
    metrics=['accuracy']
)
model.fit(x_train_falttened,y_train,epochs=5)
```

```python
model.evaluate(x_test_falttened,y_test)
```

```python
plt.matshow(x_test[0])
```

```python
x_test_predict=model.predict(x_test_falttened)
np.argmax(x_test_predict[0])
```

```python
y_predict=[]
i=0
for x in x_test_predict:
    y_predict.append(np.argmax(x_test_predict[i]))
    i=i+1
```

```python
y_predict
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predict)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
```

```python
import pandas as pd
df=pd.read_csv('insurance_data.csv')
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2,random_state=2)
y_train
```

```python
x_train_scaled=x_train.copy()
x_train_scaled['age']=x_train_scaled['age']/100
x_test_scaled=x_test.copy()
x_test_scaled['age']=x_test_scaled['age']/100
```

```python
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation="sigmoid",kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(
optimizer='adam',
    loss="binary_crossentropy",
    metrics=['accuracy']
)
model.fit(x_train_scaled,y_train,epochs=5000)
```

```python
model.evaluate(x_test_scaled,y_test)
```
