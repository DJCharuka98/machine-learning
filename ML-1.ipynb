```python
x=numpy.array([1,2,3,4,5])
y=numpy.array([10,12,14,16,18])
m_cur=b_cur=0
iterations=10000
d=0.08
n=len(x)
cost_min=10000

for t in range(iterations):
    pre_y=m_cur*x+b_cur
    par_m=2/len(x)*sum((-x)*(y-pre_y))
    par_b=2/len(x)*sum((-1)*(y-pre_y))
    cost=1/len(x)*sum((y-pre_y)**2)
    
    print("cost:{}  || m:{}  ||  b:{}".format(cost,m_cur,b_cur))
    m_cur=m_cur-d*par_m
    b_cur=b_cur-d*par_b
    if(int(cost)<=cost_min):
        cost_min=cost
        b_min=b_cur
        m_min=m_cur
print("_____________________________________")
print("minimal cost:{} slope:{} intercept:{}".format(cost_min,m_min,b_min))

    
        
    

    
```

K FOLD CROSS VALIDATION

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
from sklearn import tree
modeltree=tree.DecisionTreeClassifier()
```

```python
from sklearn import svm
modelsvm=svm.SVC()
```

```python
from sklearn.ensemble import RandomForestClassifier
modelrandom=RandomForestClassifier(n_estimators=40)
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=3)
kf
```

```python
for train,test in kf.split([1,2,3,4,5,6,7,8,9]):
    print(train,test)
```

```python
for train,test in kf.split(digits.data):
    print(train,test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds=StratifiedKFold(n_splits=3)
```

```python
for train,test in kf.split(digits.data):
    x_train,x_test,y_train,y_test=dif=digits.data[train],digits.data[test],digits.target[train],digits.target[test]
    
```

```python
tree=[]
svmm=[]
random=[]
def get_score(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    return model.score(x_test,y_test)
    
```

```python
for train,test in kf.split(digits.data):
    x_train,x_test,y_train,y_test=dif=digits.data[train],digits.data[test],digits.target[train],digits.target[test]
    tree.append(get_score(modeltree,x_train,x_test,y_train,y_test))
    svmm.append(get_score(modelsvm,x_train,x_test,y_train,y_test))
    random.append(get_score(modelrandom,x_train,x_test,y_train,y_test))
```
