```python
import string
string.punctuation
```

```python

def remove_punc(text):
    c=''
    for x in text:
        if(x not in string.punctuation):
            c=c+x
    return c
    
```

```python
df['msg_clean']=df['msg'].apply(remove_punc)
```

```python
df
```

```python
import re
def tokenize(text):
    tokens=re.split('\W+',text)
    
    return tokens

```

```python
df['msg_token']=df['msg_clean'].apply(tokenize)
```

```python
df
```

```python
def lower(tok):
    to=[]
    for x in tok:
        to.append(x.lower())
    return to
df['msg_token']=df['msg_token'].apply(lower)
df
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
#stop=stopwords.words('english')
#stop
```

```python
stop=stopwords.words('english')
stop
```

```python
def remove_stop(text):
    h=[]
    for x in text:
        if(x not in stop):
            h.append(x)
    return h
df['msg_stp']=df['msg_token'].apply(remove_stop)
```

```python
df
```

```python
import nltk
from nltk.stem import PorterStemmer
ps=PorterStemmer()
dir(ps)
```

```python
ps.stem('coding')
```

```python
ps.stem('coder')
```

```python
def stem_(text):
    v=[]
    for x in text:
        v.append(ps.stem(x))
    return v
        
```

```python
df['stem_msg']=df['msg_stp'].apply(stem_)
df
```

```python
wn=nltk.WordNetLemmatizer()
def lem_(text):
    v=[]
    for x in text:
        v.append(ps.stem(x))
    return v
```

```python
df['lem_msg']=df['msg_stp'].apply(lem_)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
## create a document-term matrix
```

```python
c=['this is a sine is','this is available for me','this is second time to contact you']
x=cv.fit(c)
print(x.vocabulary_)

```

```python
print(cv.get_feature_names())
```

```python
x=cv.transform(c)
x
```

```python
print(x.shape)
```

```python
print(x.toarray())
```

```python
b=pd.DataFrame(x.toarray(),columns=cv.get_feature_names())
b
```
